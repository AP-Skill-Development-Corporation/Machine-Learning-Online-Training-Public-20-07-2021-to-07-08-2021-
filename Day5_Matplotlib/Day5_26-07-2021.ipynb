# Matplotlib
- Importance of Visualization
    - To understand the data clearly when there is large data
    - to understand easier
    - For better and quick understanding

- pyplot Module is used to represent the data in the graphical format.
* The various types of plots available in matplotlib are
1. Line plot : Line plot is the linear graph,in which all points are join by a line.
2. Scatter Plot : it helps in identifying relation between two variables
3. Histogram : Probability Distribution
4. Bar Graph
5. Pie Chart
6. Box Plot

```python
from matplotlib import pyplot as plt
```

# Line Plot

```python
li = [10,20,30,40,50,60]
plt.plot(li)
plt.show()
```

```python
rate = [48500,48000,50000,49800,50000]
dates = [22,23,24,25,26]
plt.plot(dates,rate)
plt.xlabel("Dates ")
plt.ylabel("Rate in rs")
plt.title("Gold Rate in Vizag")
plt.show()
```

```python
import numpy as np
x = np.arange(1,15)
y = x**2
plt.plot(x,y,c = "red", marker = "o", linestyle = "-.")
plt.show()
```

```python
plt.plot(x,y,c = "black", marker = "+", linestyle = "-.", label = "x and x*2")
plt.plot(x,y*2,c = "y", marker = "d", linestyle = "-", label = "x and y*2")
plt.legend(loc = "center")
plt.show()
```

```python
x = np.linspace(0,10,40)
y = np.sin(x)
plt.plot(x,np.cos(x), label= "Cos(x)", marker = "*", c = "m")
plt.plot(x,y, label= "Sin(x)", marker = "^", c = "c")
plt.legend()
plt.show()
```

# Subplots

```python
fig, area = plt.subplots(2,3)
```

```python
area
```

```python
fig,area = plt.subplots(2,2)
x = np.linspace(1,10,20)
area[0][0].plot(x,x)
area[0][0].set_title("Line")
area[0][1].plot(x,x**2)
area[0][1].set_title("square")
area[1][0].plot(x,x**3)
area[1][0].set_title("cube")
area[1][1].plot(x,np.log(x))
area[1][1].set_title("Log")

plt.show()
```

```python
gp1 = plt.subplot2grid((3,3),(0,0),colspan = 2)
gp2 = plt.subplot2grid((3,3),(0,2),rowspan = 2)
gp3 = plt.subplot2grid((3,3),(1,0),colspan = 2, rowspan = 1)
x = np.arange(1,20)
gp1.plot(x, np.log(x))
gp2.plot(x,np.exp(x))
gp3.plot(x,np.cos(x))
plt.show()
```

# Scatter Plot

```python
x = np.arange(10,20)
y = np.random.randint(10,20,10)
plt.scatter(x,y, c = "r", marker = "^")
plt.show()
```

```python
x = np.arange(0,30)
y1 = np.random.randint(0,30,30)
y2 = np.random.randint(0,30,30)

plt.scatter(x,y1, c = "g", marker = "*", label = "Non-covid")
plt.scatter(x,y2, c= "r", marker = "^", label = "Covid")
plt.legend()
plt.show()
```

```python
plt.scatter(x,y1, c = "r", marker = "*")
plt.plot(x,y1, c = "m", linestyle= "--")
plt.show()
```

# Histograms
- It works on Frequency count of your data
- Probability Distribution

```python
li = [23,22,21,23,21,24,23,21, 24]
plt.hist(li, color = "c", width = 0.5, )
plt.xlabel("list data")
plt.ylabel("Frequency count")
plt.show()
```

```python
li = [23,22,21,23,21,24,23,21, 24]
plt.hist(li, color= "m",histtype = "step" )
plt.xlabel("list data")
plt.ylabel("Frequency count")
plt.show()
```

```python
li = [0,1,2,1,2,3,2,3,2,4,3,5,3]
plt.hist(li, color= "m",histtype = "step" )
plt.xlabel("list data")
plt.ylabel("Frequency count")
plt.show()
```

```python
li = [0,1,2,1,2,3,2,3,2,4,3,5,3,9,10,7]
plt.hist(li, color= "m",histtype = "stepfilled" )
plt.xlabel("list data")
plt.ylabel("Frequency count")
plt.show()
```

```python
x = np.random.randn(100)
plt.hist(x, histtype = "step")
```

# Bar Graph

```python
subjects = ["C","Python","C++","Java","R"]
marks = [80,98,79,76,40]
plt.bar(subjects,marks, color = ["c","m","b","y","r"] , width = 0.3)
plt.xlabel("Subject Names")
plt.ylabel("Marks outof 100")
plt.show()
```

```python
subjects = ["C","Python","C++","Java","R"]
marks = [80,98,79,76,40]
plt.barh(subjects,marks, color = ["c","m","b","y","r"])
plt.xlabel("Subject Names")
plt.ylabel("Marks outof 100")
plt.text(40,"R","Low")
plt.text(98,"Python","High")
plt.show()
```

# Pie chart

```python
plt.pie(marks, labels = subjects, colors= ["r","c","m","y","b"], radius = 1.5, autopct= "%1.2f%%")
plt.show()
```

```python
help(plt.pie)
```

# Box Plot
- Statastical plot
- max value, min value, 25% quantile and median value, 75% quantile, outliers


```python
x = np.random.randn(100)
print(x)
print(min(x))
print(max(x))
print(np.median(x))
plt.boxplot(x)
plt.show()
```
